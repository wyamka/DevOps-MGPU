```python
import pandas as pd
import numpy as np
```

```python
# Загружаем исходный датасет
df = pd.read_csv("diabetes.csv")

# Генерация случайных имен и фамилий из списка
first_names = ["Mary", "Jane", "Anna", "Susan", "Patricia", "Linda", "Abby", "Chloe", "Bertha", "Alice", "Isabel", "Jill", "Lana", "Katie", "Megan", "Nora", "Olivia", "Paula", "Rita", "Sophie", "Tina", "Ursula", "Vera", "Wendy"]

last_names = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Davis", "Miller", "Wilson", "Moore", "Taylor", "Anderson", "Thomas", "Jackson", "White", "Harris", "Martin", "Clark", "Lewis", "Robinson", "Walker", "Hall", "Allen", "Young", "Hanks"]
np.random.seed(42)
df["FirstName"] = np.random.choice(first_names, size=len(df))
df["LastName"] = np.random.choice(last_names, size=len(df))
```
